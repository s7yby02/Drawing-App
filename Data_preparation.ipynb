```python
import glob
import json
import sys
from pprint import pprint
```

```python
#store all the json files in a variable
jsonfiles = glob.glob(r'data/raw/*')
```

```python
for i in jsonfiles:
    print(i)
```

```python
for i in jsonfiles:
    with open(i,'r') as jsonfile:
        d=json.load(jsonfile)
    print(d)
    print(d['drawings'])
    print(len(d['drawings']))
    print(len(d['drawings']['car']))
    # pprint(d['drawings']['car'])
    break
```

##### Creating new data with same json file but with different names and slightly modified for easy use

```python
for i,file in enumerate(jsonfiles):
    with open(file,'r',encoding='utf-8') as jsonfile:
        d=json.load(jsonfile)
        #print(d)
        d['session']=i
        del d['student']
        path='newdata/json/'+str(i)+'.json'
        with open(path,'w') as file:
            json.dump(d,file,indent=2)
            #print(f"{i}.json file is created!")
            sys.stdout.write(f'\rCreating file {i+1}/{len(jsonfiles)} ({(i+1) / len(jsonfiles) * 100:.2f}%)')
            sys.stdout.flush()

    '''if i==5:
        break'''
```
